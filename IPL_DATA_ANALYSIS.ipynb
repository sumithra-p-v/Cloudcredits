```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
data = pd.read_csv('/content/df_batting (1).csv')
data2 = pd.read_csv('/content/df_batting (1).csv')
```

```python
data['SR'] = pd.to_numeric(data['SR'], errors='coerce')
```

```python
data.describe(include = 'number').T
```

```python
sns.pairplot(data)
plt.show()
```

```python
for i in data.select_dtypes(include = 'number').columns:
    fig = px.histogram(data, x = i)
    fig.update_layout(
    title = f'{i}',
    height = 450,
    width = 700
    )
    fig.show()
```

```python
for i in data.select_dtypes(include = 'number').columns:
    fig = px.box(data, x = i)
    fig.update_layout(
    title = f'{i}',
    height = 400,
    width = 600
    )
    fig.show()
```

```python
fig = px.scatter(data,
                 x = 'balls',
                 y = 'runs',
                 color = 'batsmanName',
                  title = 'Runs vs Balls Faced',
                  labels = {'balls':'Balls Faced', 'runs':'Runs Scored'})
fig.show()
```

```python
fig = px.bar(data,
             x = 'battingPos',
             y = 'runs',
             color = 'out/not_out',
             title = 'Runs by Batting Position',
             labels = {'battingPos':'Batting Position', 'runs':'Runs Scored', 'out/not_out':'Out/Not Out'})
fig.show()
```

```python
fig = px.box(data,
             x = 'out/not_out',
             y = 'runs',
             title = 'Runs by Out/Not Out Status',
             labels={'out/not_out':'Out/Not Out', 'runs':'Runs Scored'})
fig.show()
```

```python
fig = px.scatter(data,
                 x = 'runs',
                 y = 'SR',
                 color = 'batsmanName',
                 title = 'Strike Rate vs Runs Scored',
                 labels = {'runs':'Runs Scored', 'SR':'Strike Rate'})
fig.show()
```

```python
s = data.select_dtypes(include = 'number').corr()
fig = px.density_heatmap(s, title = 'Correlation Heatmap')
fig.update_layout(
height = 600,
width = 800
)
fig.show()
```

```python
data2['matchDate'] = pd.to_datetime(data2['matchDate'])
fig = px.line(data2,
              x = 'matchDate',
              y = 'winner',
              title = 'Matches Over Time',
              labels = {'matchDate': 'Date', 'winner': 'Winner'})
fig.show()
```

```python
data2_margin = data2.dropna(subset=['margin'])
fig = px.bar(data2_margin,
             x = 'winner',
             y = 'margin',
             title = 'Wins by Margin',
             labels = {'winner': 'Winner', 'margin': 'Margin'})
fig.show()
```
